```python
import numpy as np 
import pandas as pd 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df=pd.read_csv("netflix_titles.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
data=df[["show_id","title","director","cast","description"]]
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.fillna("", inplace=True)
```

```python
data.isnull().sum()
```

```python
data.iloc[2].cast
```

```python
data.iloc[2].director
```

```python
data.iloc[2].description
```

```python
def transform(obj):
    if isinstance(obj, str):
        return obj.replace(",","").split()
    else:
        return [obj]
    
```

```python
data["cast"]=data["cast"].apply(transform)
```

```python
data["director"]=data["director"].apply(transform)
```

```python
data["description"]=data["description"].apply(transform)
```

```python
data.head()
```

```python
data["key"]=data["cast"]+data["description"]+data["director"]
```

```python
df_2=data[["show_id","title","key"]]
```

```python
df_2.head()
```

```python
df_2["key"]=df_2["key"].apply(lambda x:" ".join(x).lower())
```

```python
df_2.head()
```

VECTORIZATION

```python
import nltk
from nltk.stem.porter import PorterStemmer 
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
df_2["key"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=8000, stop_words="english")
```

```python
vectors=cv.fit_transform(df_2["key"])
```

```python
cv.get_feature_names_out()
```

COSINE DISTANCE

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index=df_2[df_2["title"]== movie].index[0]
    movie_list=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list :
        print(df_2.iloc[i[0]].title)
```

```python
recommend("Kota Factory")
```
